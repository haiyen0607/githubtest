# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
sales = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
sales.head()
```

```python
sales['total_profit']
```

```python
sales.index += 1
plt.plot(sales.total_profit, linewidth = 2)
plt.title('Company profit per month', size = 15)
plt.xlabel('Month number', size = 10)
# x = np.arange(1,13)
y = np.arange(100000,600000,100000)
plt.xticks(range(1, 13))
plt.yticks(y)
plt.ylim(100000, 500001)
plt.xlim(0,13)
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
plt.figure(figsize=(8,5.5))
sales['total_profit'].plot(linestyle = '--', marker = 'o', color = 'red', markerfacecolor = 'black', linewidth = 3.5, markersize = 8)
plt.title('Company Sales data of last year', size = 15)
plt.xlabel('Month number', size = 13)
x = np.arange(1,13)
y = np.arange(100000,600000,100000)
plt.xticks(x)
plt.yticks(y)
plt.ylim(100000, 500001)
plt.xlim(0,13)
plt.legend(['Profit data of last year'], loc = "lower right")
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
plt.figure(figsize = (8,5.5))
plt.plot(sales.iloc[:,:6], linewidth = 4, marker = 'o', markersize = 8)
plt.title('Sales Data', size = 15)
plt.xlabel('Month number', size = 13)
x = np.arange(1,13)
y = np.arange(1000,19000,2000)
plt.xticks(x)
plt.yticks(y)
plt.legend(['Face cream Sales Data', 'Face Wash Sales Data', 'ToothPaste Sales Data', 'Bathingsoap Sales Data', 'Shampoo Sales Data', 'Moisturizer Sales Data'])
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
tooth_paste = sales[['month_number', 'toothpaste']]
tooth_paste
```

```python
tooth_paste.plot(kind = 'scatter', x = 'month_number', y = 'toothpaste', figsize = (7,5), s = 50)
plt.grid(True, linestyle = '--', which = 'both', linewidth = 1.25)
plt.xticks(range(1,13))
plt.legend(['Tooth paste Data Sales'])
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
plt.figure(figsize = (7.5,6))
x_axis = np.arange(1,13)
plt.bar(x_axis, sales.facecream, width=0.3, label = 'Face Cream Sales Data')
plt.bar(x_axis + 0.3, sales.facewash, width=0.3, label = 'Face Wash Sales Data')
plt.title('Facewash and Facecream Sales Data', size = 15)
plt.xlabel('Month Number', size = 12)
plt.ylabel('Sales units in number', size = 12)
plt.grid(True, linestyle = '--', which = 'both', linewidth = 1.25, label = 0.3)
plt.xticks(x_axis)
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/06.png?raw=1)

```python
plt.figure(figsize=(8,6))
x_axis = np.arange(1,13)
plt.bar(x_axis, sales.bathingsoap)
plt.title('bathingsoap Sales Data', size = 15)
plt.xlabel('Month Number', size = 12)
plt.ylabel('Sales units in number', size = 12)
plt.xticks(x_axis)
plt.grid(True, linestyle = '--', which = 'both', linewidth = 1.25)
plt.savefig('D:\Python\Articles\matplotlib\sales_of_bathingsoap.png', dpi=150)
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
plt.figure(figsize = (8,5))
listdata = sales['total_profit'].tolist()
labels = ['low', 'average', 'Good', 'Best']
Profit_range = [150000,175000,200000,225000,250000,300000,350000]
plt.hist(listdata, Profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar', size = 12)
plt.ylabel('Actual Profit in dollar', size = 12)
plt.title('Profit data', size = 15)
plt.legend(loc = 'upper left')
plt.xticks(Profit_range)
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
listmonth = sales['month_number'].tolist()
label = ['FaceCream', 'FaceWash', 'ToothPaste', 'BathingSoap', 'Shampoo', 'Moisturizer']
sale_data = [sales['facecream'].sum(), sales['facewash'].sum(), sales['toothpaste'].sum(), sales['bathingsoap'].sum(), sales['shampoo'].sum(), sales['moisturizer'].sum()]
plt.axis('equal')
plt.pie(sale_data, labels = label, autopct='%1.1f%%' )
plt.title('Sales data')
plt.legend(loc = 'lower right')
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
listsoap = sales['bathingsoap'].tolist()
listwash = sales['facewash'].tolist()
listmonth = sales['month_number'].tolist()
data, axis = plt.subplots(2, sharex = True)
axis[0].plot(listmonth, listsoap, color = 'black', linewidth = 3, marker = 'o')
axis[1].plot(listmonth, listwash, color = 'red', linewidth = 3, marker = 'o')
axis[0].set_title('Sales data of a facewash', size = 13)
axis[1].set_title('Sales data od a facewash', size = 13)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(listmonth)
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
plt.figure(figsize=(7.5,5))
listmonth  = sales['month_number'].tolist()

faceCremSalesData   = sales['facecream'].tolist()
faceWashSalesData   = sales['facewash'].tolist()
toothPasteSalesData = sales['toothpaste'].tolist()
bathingsoapSalesData   = sales['bathingsoap'].tolist()
shampooSalesData   = sales['shampoo'].tolist()
moisturizerSalesData = sales['moisturizer'].tolist()

plt.plot([],[],color='m', label='face Cream', linewidth=5)
plt.plot([],[],color='c', label='Face wash', linewidth=5)
plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(listmonth, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])


plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
